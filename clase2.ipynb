# Problema, un modelo de ahorro
Suponga que depositamos $\$ 5000 $ en una cuenta de ahorros con intereses devengados a una tasa de capitalización continua del $ 2\% $. Si dejamos que $A (t)$ denote la cantidad de dinero en la cuenta en el tiempo t, entonces la ecuación diferencial para $A$ es

$\frac{dA}{dt}=0.02A$

* ¿Encuentre la solución análitica 
* Suponiendo que el interes nunca cambia, cual sería el dinero que tendría despues de 10 años

Si vemos que al cabo de 10 años, tenemos una buena cantidad de dinero, por ende decidimos retirar cada año la cantidad de $ \$ 500$, la ecución diferencial cambia de la siguiente forma, cuando $t>10$$ 

$\frac{dA}{dt}=0.02A-500$

De esta forma podemos expresar la ecuación para todo tiempo

$ \frac{dA}{dt}=\begin{cases}0.02A & \text{si } t<10,\\
0.02A-500& \text{si } t>10.\end{cases} $


¿Al cabo de 12 años cuanto dinero tendremos?


# Campo de direcciones
Encontrar una solución análitica de una ecuación diferencial no es tan sencillo, pero si podemos describir el comportamiento de la solución de esta usando métodos cualitativos, para el caso de una ecuación diferencial de primer orden podemos usar los campos de direcciones el cual puede ayudar a describir el comportamiento de la solución.


## Ejemplo
Un campo de direcciones, el siguiente coódigo nos muestra el campo de direcciones de la ecuación diferencial ordinaria 
$y'=2\frac{y}{x}$

```python


import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np


# Ecuacion Diferencial: dy/dx=f(x,y)
        

r=2.0
K=3.0
def f( y,x): 
    return r*y*(1-y/K)
    #return -y+np.sin(4*x)



# Campo de pendientes
N = 30  # numero de pendientes a graficar
xi = -1.3
xf = 5  # dominio
x = np.linspace(xi, xf, N)
y = np.linspace(-0.1, K+0.5*K, N)
X, Y = np.meshgrid(x, y)
U = 1
V = f(Y,X)
U2, V2 = U/np.sqrt(U**2+V**2), V/np.sqrt(U**2+V**2)
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientes: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')
```

# Solución numérica de la ecuación diferencial 


```python
# initial condition
y0 = [0, 0.1, 0.2, 0.6]

# time points
t = np.linspace(0, xf)

# solve ODE
for yi in y0:
    y = odeint(f, yi, t, rtol=1.0E-8)
    plt.plot(t, y)

# plot results

plt.xlabel('time')
plt.ylabel('y(t)')
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientesny solución de la ecuación: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')
plt.show()

```

# Definición
Una ecuación diferencial de la forma 
$ y'=f(y) $ se llama autónoma de primer orden. Note que la función $ f$ sólo depende de la variable $ y$

# Definición
Una ecuación diferencial de la forma 

$ \frac{dy}{dx}=f(x,y),\text{ sujeto a } y(0)=y_0 $

se llama un problema de valor inicial

# Teorema de Existencia 



Suponga que $ f(x,y) $  y $ \frac{\partial f}{\partial y}$ son funciones continuas en el intervalo $ \{ (x,y):a< x < b> \text{ y } c<y<d \} $ en el plano $ x-y $.  Si $ (x_0,y_0) $ son un punto del rectángulo, entonces, existe un $ \epsilon >0 $ y una función $ y(x) $ definida en el intervalo $ (x_0-\epsilon, x_0+\epsilon) $, que resuelve el porblema de valor inicial 

$ \frac{dy}{dx}=f(x,y),\text{ sujeto a } y(0)=y_0 $




# Teorema de unicidad de la solución


Suponga que $ f(x,y) $ es una función continua en el intervalo $ \{ (x,y):a< x < b> \text{ y } c<y<d \} $ en el plano $ x-y $.  Si $ (x_0,y_0) $ es un punto del rectángulo, si $ y_1 $ y $ y_2 $ son soluciones del porblema de valor inicial 

 **$ \frac{dy}{dx}=f(x,y),\text{ sujeto a } y(0)=y_0 $**

 para todo $ t $   en el intervalo $ (x_0-\epsilon, x_0+\epsilon) $, para algún $ \epsilon>0 $  entonces 
 
 $ y_1=y_2 $  en dicho intervalo
