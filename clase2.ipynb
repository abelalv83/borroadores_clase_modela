# Campo de direcciones

Un campo de direcciones, el siguiente coódigo nos muestra el campo de direcciones de la ecuación diferencial ordinaria 
$y'=2\frac{y}{x}$

```python


import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np


# Ecuacion Diferencial: dy/dx=f(x,y)
        

r=2.0
K=3.0
def f( y,x): 
    return r*y*(1-y/K)
    #return -y+np.sin(4*x)



# Campo de pendientes
N = 30  # numero de pendientes a graficar
xi = -1.3
xf = 5  # dominio
x = np.linspace(xi, xf, N)
y = np.linspace(-0.1, K+0.5*K, N)
X, Y = np.meshgrid(x, y)
U = 1
V = f(Y,X)
U2, V2 = U/np.sqrt(U**2+V**2), V/np.sqrt(U**2+V**2)
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientes: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')
```

# Solución numérica de la ecuación diferencial 


```python
# initial condition
y0 = [0, 0.1, 0.2, 0.6]

# time points
t = np.linspace(0, xf)

# solve ODE
for yi in y0:
    y = odeint(f, yi, t, rtol=1.0E-8)
    plt.plot(t, y)

# plot results

plt.xlabel('time')
plt.ylabel('y(t)')
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientesny solución de la ecuación: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')
plt.show()

```
