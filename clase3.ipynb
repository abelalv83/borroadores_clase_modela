2. Los siguientes problemas estan relacionados con el crecimiento y decrecimiento lógistico

    a. Resuleva le problema de valor inicial $y'=y(1-y/20) $ con $y(0)=5, 10,2, 30$

    b. Modique el valor de r y K en el siguiente programa para gráficar el campo de pendientes 

    

```python
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import numpy as np


# Ecuacion Diferencial: dy/dx=f(x,y)
        

r=2.0
K=3.0
def f( y,x): 
    
    return r*y*(1-y/K)
    



# Campo de pendientes
N = 30  # numero de pendientes a graficar
xi = 0
xf = 10  # dominio
x = np.linspace(xi, xf, N)
y = np.linspace(-0.1, K+0.5*K, N)
X, Y = np.meshgrid(x, y)
U = 1
V = f(Y,X)
U2, V2 = U/np.sqrt(U**2+V**2), V/np.sqrt(U**2+V**2)
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(r'Campo de pendientes: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')

```

        c. Trace las curvas solución 

```python
# initial condition
y0 = [0.3, 0.4, 0.5, 4, 3.6]

# time points
t = np.linspace(0, xf)

# solve ODE
for yi in y0:
    y = odeint(f, yi, t, rtol=1.0E-8)
    plt.plot(t, y)

# plot results

plt.xlabel('time')
plt.ylabel('y(t)')
plt.quiver(X, Y, U2, V2, color='k')
plt.grid(True)
plt.title(
    r'Campo de pendientesny solución de la ecuación: $\frac{dy}{dx}=ry(1-\frac{y}{K})$')
```
